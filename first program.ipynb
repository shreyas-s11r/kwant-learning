```python
# Create a 1D lattice
lat = kwant.lattice.chain()

# Create an empty system
sys = kwant.Builder()

# Add five atoms
for i in range(5):
    sys[lat(i)] = 0

# Connect neighboring atoms
for i in range(4):
    sys[lat(i), lat(i + 1)] = -1

# Finalize the system
fsys = sys.finalized()

print(fsys.hamiltonian_submatrix())
```
